#### 4. Write a function that prints any word that appeared in the last 20% of a text that had not been encountered earlier. Use text1 from nltk.book to call this function.

```python
def last_20(text):
    cut = int(0.8 * len(text)) 
    first_part, second_part = text[:cut], text[cut:]
    unique_words_first = set(first_part)
    unique_words_second = set(second_part)
    return [word for word in unique_words_second if word not in unique_words_first]
```

```python
from nltk.book import text1
```

```python
text = [w for w in text1]
```

```python
last_20(text)
```

#### 6. Write a function shorten(word, n) to process a text (“big big big world today tomorrow good Today good”), omitting the n most frequently occurring words of the text. You should use w.lower() to normalize the text first. Please call this shorten function.

```python
def shorten(word, n):
    text = [w.lower() for w in nltk.word_tokenize(word)]
    fdist = set(w for (w, _) in nltk.FreqDist(text).most_common(n))
    return set(w for w in text if w not in fdist)
```

```python
text =  ('big big big world today tomorrow good Today good')
```

```python
shorten(text, 3)
```

#### 10. Choose your own webpage (in html format) and output the 20 most common words in this web page. You should use w.lower() to normalize the text. Please get rid of stop words, numbers, and punctuations. Please define a function and call that function.

```python
from urllib import request
from bs4 import BeautifulSoup
from nltk.corpus import stopwords


def freq_words(url, n):
    html = request.urlopen(url).read().decode('utf8')
    text = BeautifulSoup(html, 'html.parser').get_text()
    fd = nltk.FreqDist(word.lower() for word in nltk.word_tokenize(text) if word.isalpha() and word.lower() not in nltk.corpus.stopwords.words('english'))
    return [word for (word, _) in fd.most_common(n)] 
```

```python
mets = 'https://sports.yahoo.com/mets-aaron-loup-free-agent-145032946.html'
```

```python
freq_words(mets, 20)
```
