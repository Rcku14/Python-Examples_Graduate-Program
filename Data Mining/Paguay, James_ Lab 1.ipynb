```python
data_url = "http://covidtracking.com/api/states/daily.csv"
covid = pd.read_csv(data_url, parse_dates = ['date'])
## Problem 5 Solution
covid
```

```python
## Problem 5.1 Solution
new_covid = covid[['state', 'recovered']]
new_covid.dropna(axis = 0, how = "any")
recovered_max = new_covid.recovered.idxmax()
high_state = new_covid.loc[recovered_max]
high_state
```

## Alternate Solution to Problem 5.1 Solution

```python
covid[covid['recovered']==covid['recovered'].max()]['state']
```

```python
## Plotting Most Values
lastest_date = covid['date'].max()
most_recent = covid[covid['date'] == lastest_date]
plot_order = most_recent.groupby('state')['recovered'].max().sort_values(ascending = False).index
fig, ax = plt.subplots(figsize = (28,8))
sns.barplot(x = 'state', y = 'recovered', data = most_recent, order = plot_order)
plt.show()
```
